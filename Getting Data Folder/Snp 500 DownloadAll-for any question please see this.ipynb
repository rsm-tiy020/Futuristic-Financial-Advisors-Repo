## After check Weki, we find that BRK.B is Berkshire Hathaway Inc. and BF.B is Brown-Forman Corporation, they use - instead of . in their tickers

```python
yf.download('BF-B')['Adj Close']
```

```python
yf.download('BRk-B')['Adj Close']
```

## So we change the tickers to BRK-B and BF-B

```python
for i in range(len(tickers)):
    if tickers[i] == 'BF.B':
        tickers[i] = 'BF-B'
    elif tickers[i] == 'BRK.B':
        tickers[i] = 'BRK-B'
```

```python
'BF-B' in tickers
'BRK-B' in tickers
```

## All Done now, change the code to whatever period you want
