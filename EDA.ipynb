### Stock Movement Plot

```python
import matplotlib.pyplot as plt
```

```python

# Create DataFrame
df = pd.DataFrame(data2, index=data2.index)

# Transpose the DataFrame to get dates as index
df_transposed = df.transpose()
df_transposed.head()
```

```python
# Select key dates for x-axis labels
def select_key_dates(dates, freq='YS'):
    dates = pd.to_datetime(dates)
    key_dates = pd.date_range(start=dates.min(), end=dates.max(), freq=freq)
    key_dates = key_dates.union([dates.min(), dates.max()])
    return key_dates

key_dates = select_key_dates(df_transposed.index)
key_dates
```

```python
# Ensure only date columns are included
df_transposed.index = pd.to_datetime(df_transposed.index, errors='coerce')
# Convert all values to numeric, ensuring any non-numeric values are handled
df_transposed = df_transposed.apply(pd.to_numeric, errors='coerce')

# Plot the data
plt.figure(figsize=(12, 12))

for column in df_transposed.columns:
    plt.plot(df_transposed.index, df_transposed[column], label=column)

# Set x-axis with key dates
ax = plt.gca()
ax.set_xticks(key_dates)
ax.set_xticklabels(key_dates.strftime('%Y-%m-%d'))

# Adding stock names at the end of each line
for column in df_transposed.columns:
    # Get the last valid value and its date
    last_valid_value = df_transposed[column].values[np.isfinite(df_transposed[column].values)][-1]
    last_valid_date = df_transposed.index[np.isfinite(df_transposed[column].values)][-1]
    plt.text(last_valid_date, last_valid_value, column, fontsize=9)

plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend().set_visible(False)  # Hide the legend to avoid duplication
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### Moving Averages

```python
# Calculate moving averages
window_size = 20
df_moving_avg = df_transposed.rolling(window=window_size).mean()

plt.figure(figsize=(12, 6))
for column in df_transposed.columns:
    plt.plot(df_transposed.index, df_transposed[column], label=column)
    plt.plot(df_moving_avg.index, df_moving_avg[column], linestyle='--')
    
    # Adding stock names at the end of each line
    last_valid_value = df_transposed[column].dropna().iloc[-1]
    last_valid_date = df_transposed.index[df_transposed[column].notna()][-1]
    plt.text(last_valid_date, last_valid_value, column, fontsize=9, verticalalignment='center')

plt.title('Stock Prices and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend().set_visible(False)  # Hide the legend to avoid duplication
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### Volatility Analysis

```python
rolling_std = df_transposed.rolling(window=20).std()

plt.figure(figsize=(12, 6))
for column in rolling_std.columns:
    plt.plot(rolling_std.index, rolling_std[column], label=column)
    
    # Adding stock names at the end of each line
    last_valid_value = rolling_std[column].dropna().iloc[-1]
    last_valid_date = rolling_std.index[rolling_std[column].notna()][-1]
    plt.text(last_valid_date, last_valid_value, column, fontsize=9, verticalalignment='center')

plt.title('Rolling Standard Deviation (Volatility)')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend().set_visible(False)  # Hide the legend to avoid duplication
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### Correlation Analysis

```python
import seaborn as sns

correlation_matrix = df_transposed.corr()

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Stock Price Correlation Matrix')
plt.show()
```

```python
# Use a mask to display only the lower triangle of the correlation matrix
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Draw the heatmap with the mask
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={"shrink": .8}, linewidths=0.5)

plt.title('Stock Price Correlation Matrix')
plt.show()
```

### Returns Analysis

```python
returns = df_transposed.pct_change().dropna()

plt.figure(figsize=(12, 6))
for column in returns.columns:
    plt.plot(returns.index, returns[column], label=column)
    
    # Adding stock names at the end of each line
    last_valid_value = returns[column].dropna().iloc[-1]
    last_valid_date = returns.index[returns[column].notna()][-1]
    plt.text(last_valid_date, last_valid_value, column, fontsize=9, verticalalignment='center')

plt.title('Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend().set_visible(False)  # Hide the legend to avoid duplication
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Determine which stocks to label
n_labels = 5  # Number of stocks to label from the center
top_labels = returns.iloc[-1].nlargest(2)  # Label top performers
bottom_labels = returns.iloc[-1].nsmallest(2)  # Label bottom performers
# center_labels = returns.iloc[-1].sort_values().iloc[len(returns.columns) // 2 - n_labels // 2 : len(returns.columns) // 2 + n_labels // 2]
# center_labels = returns.iloc[-1].sort_values().iloc[[len(returns.columns) // 2]]  # Label the center stock
# Select 3 equidistant center stocks
# middle_indices = np.linspace(0, len(returns.columns) - 1, 7, dtype=int)[2:5]
# center_labels = returns.iloc[-1].sort_values().iloc[middle_indices]
# Select 2 equidistant center stocks
sorted_returns = returns.iloc[-1].sort_values()
center_indices = np.linspace(0, len(sorted_returns) - 1, 10, dtype=int)[1:3]
center_labels = sorted_returns.iloc[center_indices]
```

```python
plt.figure(figsize=(12, 6))
for column in returns.columns:
    plt.plot(returns.index, returns[column], label=column)


# Adding stock names at the end of each line
for column in returns.columns:
    last_valid_value = returns[column].dropna().iloc[-1]
    last_valid_date = returns.index[returns[column].notna()][-1]
    if column in top_labels.index or column in bottom_labels.index or column in center_labels.index:
        plt.text(last_valid_date, last_valid_value, column, fontsize=9, verticalalignment='center')

plt.title('Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend().set_visible(False)
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
