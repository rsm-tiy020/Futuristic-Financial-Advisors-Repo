```python
import pandas as pd
import yfinance as yf
```

```python
sp500url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data_table = pd.read_html(sp500url)
```

```python
tickers = data_table[0]['Symbol'].tolist()
tickers
```

```python
for i in range(len(tickers)):
    if tickers[i] == 'BF.B':
        tickers[i] = 'BF-B'
    elif tickers[i] == 'BRK.B':
        tickers[i] = 'BRK-B'
```

```python
snp_prices = yf.download(tickers, start='2014-05-01', end='2024-05-01')
```

```python
snp_prices
```

```python
snp_prices.to_csv('snp_prices_10yrs.csv')
```

```python
import zipfile

def compress_csv_zip(input_file, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zf:
        zf.write(input_file, arcname=input_file)

# Example usage
compress_csv_zip('snp_prices_10yrs.csv', 'snp_prices_10yrs.zip')
```
